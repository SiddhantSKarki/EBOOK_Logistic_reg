```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv("presidents.csv", index_col=0)
```

```python
#cleaning President's Name column
```

```python
pattern = "(?P<first>^[\w]*)(?:.* )(?P<last>[\w]*$)"
```

```python
names = df["President"].str.extract(pattern)
```

```python
df = pd.concat([names, df], axis=1)
del(df["President"])
```

```python
#cleaning the DOB and DOD
born_dates = df["Born"].str.extract("(?P<born>[\w]{3} [0-9]{1,2}, [0-9]{4})")
dead_dates = df["Died"].str.extract("(?P<died>[\w]{3} [0-9]{1,2}, [0-9]{4})")
df = pd.concat([df,born_dates, dead_dates], axis=1)
```

```python
df["born"]= pd.to_datetime(df["born"])
df["died"]= pd.to_datetime(df["died"])
```

```python
del([df["Born"], df["Died"]])
```

```python
#cleaning age column
pattern_age = "(?P<years>^[0-9]{2})(?:.{8})(?P<days>[0-9]{1,3})"
df["Age"].str.extract(pattern_age)
```
