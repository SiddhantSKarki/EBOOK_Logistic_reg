```python
import pandas as pd
```

```python
df = pd.read_csv("presidents.csv", index_col = 0)
```

```python
# Data Cleaning starts
seg_1 = df["president"].str.extract("(?P<first_name>^[\w]*)(?:.* )(?P<last_name>[\w]*$)")
```
